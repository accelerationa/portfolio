```python
from pandas import *
from numpy import *
from operator import *
import statsmodels.formula.api as sm
```

```python
Tickers=read_table('Ticker.txt')
Ticker=transpose(array(Tickers)).tolist()
Ticker=Ticker[0]
```

```python
from pandas_datareader import data as pdr
import datetime
ls_key = 'Close'        #从下下来的数据里面提取要的信息，这个可以改成其他数据
start = datetime.datetime(2015,6,1)     #设置起止时间，也可以变
end = datetime.datetime(2017,1,1)
f =pdr.DataReader('ITB','morningstar',start,end)       #这几句我是为了先创建一个DataFrame，先拿一支股票来跑
data={'ITB': array(DataFrame(f[ls_key])).tolist()}
dataFrame=DataFrame(data)
Stock_n=Ticker[:]
for i in range(0,len(Stock_n)):
    try:
        f =pdr.DataReader(Stock_n[i],'morningstar',start,end)  #这个是下数据的function，每一次下一只股票的，本来可以一起下，但是我为了把股票做成一个列表，就一次下一个，然后循环跑
        data=array(DataFrame(f[ls_key])).tolist()
        dataFrame[Stock_n[i]]=DataFrame(data)
        print(i,'is done')
    except:
        continue
dataFrame
```

```python
del dataFrame['ITB']
ReturnData=dataFrame.pct_change(1)
```

```python
CumulativeReturn=pandas.rolling_apply(ReturnData,90,lambda x: prod(1+x)-1)
```

```python
Return_list=array(ReturnData).tolist()
```

```python
CumulativeReturn_list=array(CumulativeReturn).tolist()
```

```python
Price_list=array(dataFrame).tolist()
```

```python
List_of_cross_sections=list()
for i in range(0,len(Return_list)):
    X=Return_list[i]
    Y=CumulativeReturn_list[i]
    W=Price_list[i]
    A=list()
    for j in range(0,len(Return_list[i])):
        Z=[Ticker[j],Y[j],X[j],W[j]]
        if isnan(Y[j])==False:
            A.append(Z)
    if A!=[]:
        List_of_cross_sections.append(A)
for i in List_of_cross_sections:
    i=i.sort(key=lambda elem: elem[1])
List_of_cross_sections   # TICKER, CUMULATIVE RETURN, RETURN, PRICE
```

```python
Number_of_stocks=len(Stock_n)//100
X=List_of_cross_sections[0]
Y=list()
Portfolio=list()
for n in range(1,Number_of_stocks+1):
    Y.append(X[-n])
Portfolio.append(Y)
for i in range(1,len(List_of_cross_sections)):
    Y=list()
    X=List_of_cross_sections[i]
    if i%5==0:
        for n in range(1,Number_of_stocks+1):
            Y.append(X[-n])
        Portfolio.append(Y)
    else:
        Z=Portfolio[i-1]
        for z in Z:
            for x in X:
                if z[0]==x[0]:
                    Y.append(x)
    Portfolio.append(Y)
Portfolio    # TICKER, CUMULATIVE RETURN, RETURN, PRICE
len(Portfolio)
len(Portfolio[0])
```

```python
Portfolio_return=list()
for i in Portfolio:
    Return=0
    count=0
    for j in i:
        Return=Return+j[2]
        count=count+1
    Portfolio_return.append(Return/count)
Portfolio_return
```
